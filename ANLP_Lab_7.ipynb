## **Task1:** Summarization using BART

I did follow the instructions on the lab file "CAI451-Lab7.pdf" to create python code to process different prompt “user_prompt”

**Citation and all rights for the *code*  reserved to Professor *Badr Aldaihani***

```python
# Load the pre-trained BART model and tokenizer for summarization
model_name = "facebook/bart-large-cnn"
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)
```

```python
# Function to summarize the text
def summarize_text(text, max_length=150, min_length=30, num_beams=4):

    # Encode the input text
    inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)

    # Generate summary using beam search
    summary_ids = model.generate(inputs, max_length=max_length, min_length=min_length,
                                  num_beams=num_beams, length_penalty=2.0, early_stopping=True)
    # Decode and return the summary
    return tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
# Example usage with a user prompt
if __name__ == "__main__":
    user_prompt = """Princess Nourah bint Abdulrahman University is one of the successes of
    the care and attention that has been directed toward women's higher education. The efforts
    began in 1970, when the General Presidency for Girls' Education laid the cornerstone of the first
    educational college for girls. Since then, it has established a further 102 colleges that range from
    universities to intermediate and community colleges, are distributed across 72 Saudi cities, and
    comprise 600,000 female students. In Riyadh alone, six colleges were founded: College of
    Languages, College of Sciences, College of Teacher Education, College of Social Work,
    College of Home Economics, and College of Arts. In 2006, a royal decree was issued that
    established the first university for girls in Riyadh, under the supervision of the Ministry of
    Higher Education and with the aim of serving the development and progress of the Kingdom."""

    # Get the summary
    summary = summarize_text(user_prompt)
    print("Original Text:\n", user_prompt)
    print("\nSummary:\n", summary)
```

## **Task2:** Response generation using OpenAI's GPT

I did follow the instructions on the lab file "CAI451-Lab7.pdf" to create python code to process different prompt

**Citation and all rights for the *code*  reserved to Professor *Badr Aldaihani***

```python
# Function to run a prompt using OpenAI's GPT
def run_prompt(prompt, model="gpt-3.5-turbo", max_tokens=100, temperature=0.7, top_p=0.9):
    try:
        response = openai.ChatCompletion.create(
            model=model,
            messages=[{"role": "user", "content": prompt}],
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
        )
        return response.choices[0].message['content'].strip()
    except Exception as e:
        return f"An error occurred: {e}"
```

```python
# Example usage with a prompt
if __name__ == "__main__":

  prompt = "As an environmental scientist, explain the role of urban green spaces in improving air quality."
  response = run_prompt(prompt)
  print("Response:\n", response)
```

## **Task3:** Chat with GPT

```python
# Function to run a chat-based prompt using OpenAI's GPT
def chat_with_gpt(prompt):

    try:
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "user", "content": prompt}
            ]
        )
        return response["choices"][0]["message"]["content"]

    except Exception as e:
        return f"An error occurred: {e}"
```

```python
# Example usage with user input
if __name__ == "__main__":

    user_prompt = input("Enter your prompt: ")
    response = chat_with_gpt(user_prompt)
    print("ChatGPT response:", response)

    # The input that I have used:
    '''
    As a historian,
    discuss the impact of social media on modern revolutions and movements.
    How has it changed the way people organize,
    communicate, and enact change?
    '''
```

```python
# Load LLaMA model and tokenizer from HuggingFace
model_name = "meta-llama/Llama-3.2-1B"
model = AutoModelForCausalLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
# Define the prompt
prompt = "As an AI Expert, explain the benefits of artificial intelligence in Cyber Security."
```

```python
# Tokenize the prompt
inputs = tokenizer(prompt, return_tensors="pt")
```

```python
# Generate the response
response = model.generate(
    **inputs,
    max_length=150,  # Limit the response length
    num_return_sequences=1,  # Generate one response
    temperature=0.8,  # Increase randomness for creativity
    top_p=0.9,  # Focus on high-probability responses
    pad_token_id=tokenizer.eos_token_id  # Handle padding
)
```

```python
# Decode the generated text
output = tokenizer.decode(response[0], skip_special_tokens=True)
print(output)
```
